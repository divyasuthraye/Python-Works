Exercise 1: Write a simple program to simulate the operation of the
grep command on Unix. Ask the user to enter a regular expression and
count the number of lines that matched the regular expression:<br>

$ python grep.py<br>
Enter a regular expression: ^Author<br>
mbox.txt had 1798 lines that matched ^Author<br>


```python
import re
hand = open("Test.txt")
a = input("Enter a regualar expression:")
count = 0
for line in hand:
    line = line.rstrip()
    x = re.findall(a, line)
    if len(x) > 0:
        count = count + 1
print("Test.txt had " + str(count) +" lines that matched:",a)
```

Exercise 2: Write a program to look for lines of the form:
New Revision: 39772<br>
Extract the number from each of the lines using a regular expression
and the findall() method. Compute the average of the numbers and
print out the average.

```python
import re
file =  input("Enter file: ")
sum = 0
count = 0
pattern = 'New Revision: ([0-9].*)'
with open(file,'r') as f:
    lines = f.readlines()
    for line in lines:
        num = re.findall(pattern, line)
        if(num):
            sum += float(num[0])
            count += 1
print(sum/count)
```

Exercise 1: Change the socket program socket1.py to prompt the user
for the URL so it can read any web page. You can use split('/') to
break the URL into its component parts so you can extract the host
name for the socket connect call. Add error checking using try and
except to handle the condition where the user enters an improperly
formatted or non-existent URL.

```python
import socket
mysocket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
url = input("Enter the url:")
try:
    new_url = url.split('/')[1]
    mysocket.connect((new_url, 80))
    cmd = 'GET http:/' + new_url + ' HTTP/1.0\r\n\r\n' 
    mysocket.send(cmd.encode())
    while True:
        data = mysocket.recv(512)
        if len(data) < 1:
            break
        print(data.decode(),end='')
    mysocket.close()
except(e):
    print(e)
```
